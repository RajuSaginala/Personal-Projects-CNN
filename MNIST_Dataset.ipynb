```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
```

This is a dataset of 60,000 28x28 grayscale images of the 10 digits, along with a test set of 10,000 images.

Import data

```python
 (X_train, y_train), (X_test, y_test)=tf.keras.datasets.mnist.load_data(path='mnist.npz')
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
pd.set_option('max_columns',None)
pd.DataFrame(X_train[0])
```

```python
plt.imshow(X_train[1001],cmap='gray')
```

```python
y_train[1001]
```

4.Data Preprocessing

```python
#standard deviation shld be less
X_train.std(),X_test.std()
```

```python
from keras import Sequential #create a pipeline.Starting point to build the model
from keras.layers import Flatten,Dense
```

5.1 Build your architecture

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28,1))) # converting into 1D 
#example 28*28 image is input then in input layer 784 neurons are created as 28*28=784 flatten
model.add(Dense(units=150,activation='relu'))
#model.add(Dense(units=50,activation='relu'))
model.add(Dense(units=10,activation='softmax'))
#units=no of neurons,activation function to be applied for each neuron
model.summary()
```

5.2 model compile

```python
model.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
```

6. Model Training

```python
model_training=model.fit(X_train,y=y_train,batch_size=32,epochs=20,validation_data=(X_test,y_test))
```

7. Model Evaluation

```python
model.evaluate(x=X_test,y=y_test)
```

8. Model Prediction

```python
y_pred=model.predict(X_test)
pd.DataFrame(y_pred)
```

9. Model performance

```python
model_training.history['loss']
```

```python
plt.figure(figsize=(16,6))
plt.plot(model_training.history['loss'],label='Traning Loss')
plt.plot(model_training.history['val_loss'],label='Validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Model Preformance between Epcohs Vs Loss',size=25)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
plt.plot(model_training.history['accuracy'],label='Traning accuracy')
plt.plot(model_training.history['val_accuracy'],label='Validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('Model Preformance between Epcohs Vs Accuracy',size=25)
plt.legend()
```
